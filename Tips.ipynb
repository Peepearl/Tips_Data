# Data visualization using seaborn in python

### introduction to seaborn

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn 
%matplotlib inline
```

```python
df =pd.read_csv(r"C:\Users\Patience\Desktop\Data\tips.csv")
```

```python
df.head(5)
```

```python
df.tail(5)
```

```python
#use swarmplot to visualize tip observation and amounts by day of the weeks
seaborn.swarmplot(x = "day", y = "tip", data = df)
seaborn.set_style("whitegrid")
plt.show()
```

```python
seaborn.swarmplot(x =df["tip"])
seaborn.set_style("darkgrid")
plt.show()
```

```python
#color points by category:create customised palette
gender_palette = ["#A833FF", "#FFAF33"]
seaborn.swarmplot(x = "day", y = "tip", hue = "sex", palette = gender_palette, data = df)
plt.show()
```

```python
#draw facet grid based on the variables
seaborn.violinplot(x = df["tip"], color = "gold")
plt.show()
```

```python
#draw facet grid based on the sex
seaborn.violinplot(x = "sex", y = "total_bill", data=df)
plt.show()
```

```python
#draw facet grid based on tip variable using histogram
fg = seaborn.FacetGrid(df, col = "time", row="sex")
fg = fg.map(plt.hist, "tip", color= "tomato")
```

```python
fg = seaborn.FacetGrid(df, col = "time", row="sex")
fg = fg.map(plt.scatter, "total_bill", "tip", color= "floralwhite", edgecolor="hotpink")
```

```python
x = seaborn.FacetGrid(df, col = "time", hue = "sex")
x = x.map(plt.scatter, "total_bill", "tip")
x = x.add_legend()
```
